<h1>2. Morfologia Matemática (Erosão e dilatação)</h1>

```python
import cv2
import numpy as np
import scipy.fftpack as fp
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgs/saitama.jpg', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12,8))
plt.title('imagem Original')
plt.imshow(img, cmap='gray')
```

<p>Uso da função cv2.Canny para obter a borda da imagem acima, e depois é feito a dilatação, com um elementos estruturante 5x5:</p>

```python
canny = cv2.Canny(img,100,200, 15) #Detector de borda de Canny

ele = np.ones((5,5),np.uint8)

dilate = cv2.dilate(canny, ele, 1)

plt.figure(figsize=(15,20))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('Detector de borda de Canny')
plt.imshow(canny, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('dilatação')
plt.imshow(dilate, 'gray')
```

<p>Erosão aplicado a imagem abaixo:</p>

```python
letras = cv2.imread('imgs/letras.jpg', cv2.IMREAD_GRAYSCALE)

#ele = cv2.getStructuringElement(cv2.MORPH_CROSS,(2, 2))
ele = np.ones((4,4),np.uint8)

erode = cv2.erode(letras, ele, 1)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('imagem original')
plt.imshow(letras, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('erosão')
plt.imshow(erode, 'gray')
```

<p>Abertura é uma erosão seguido de dilatação (utilizando o mesmo elemento estruturante) sendo útil na remoção de ruído da imagem, como pode ser observado no exemplo abaixo, dilatação seguida de erosão chama-se fechamento:</p>

```python
img2 = cv2.imread('imgs/ruido.png', cv2.IMREAD_GRAYSCALE)
ele = cv2.getStructuringElement(cv2.MORPH_CROSS,(5, 5))

data = cv2.erode(img2, ele)
data = cv2.dilate(data, ele)

data = cv2.bitwise_not(data) #inverte a imagem pra filtrar a parte interna

data = cv2.erode(data, ele)
data = cv2.dilate(data, ele)

data = cv2.bitwise_not(data)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('original')
plt.imshow(img2, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('filtrada')
plt.imshow(data, 'gray')
```

<p>Erosão e dilatação podem ser usados para detecção de bordas, como pode ser visto nos exemplos abaixo, onde as imagens foram dilatadas e posteriormente subtraidas as imagens originais sobrando somente as bordas :</p>

```python
img3 = cv2.imread('imgs/picapau.jpg', cv2.IMREAD_GRAYSCALE)

ele = cv2.getStructuringElement(cv2.MORPH_CROSS,(5, 5))
dilate = cv2.dilate(img3, ele)

contorno = dilate - img3

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('original')
plt.imshow(img3, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('filtrada')
plt.imshow(contorno, 'gray')
```

```python
img4 = cv2.imread('imgs/pinguins.jpg', cv2.IMREAD_GRAYSCALE)
ele = cv2.getStructuringElement(cv2.MORPH_CROSS,(8, 8))
dilate = cv2.dilate(img4, ele)

contorno = dilate - img4

contorno = cv2.bitwise_not(contorno)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('original')
plt.imshow(img4, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('filtrada')
plt.imshow(contorno, 'gray')
```
