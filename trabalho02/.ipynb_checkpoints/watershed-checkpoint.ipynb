<h1>5. Watershed</h1>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
```

<p>Carregando imagem composta por moedas de tamanhos diferentes e em contato uma com as outras</p>

```python
img = cv2.imread('imgs/coins.png')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,8))
plt.title('imagem Original')
plt.imshow(img, cmap='gray')
```

<p>A imagem é linearizada através do método de Otsum, logo a seguir são removido os ruídos usando fechameto morfológico com um kernel 2x2 e 3 interações:</p>

```python
ret, thresh = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(thresh,cv2.MORPH_CLOSE, kernel, iterations = 3)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('imagem com ruído')
plt.imshow(thresh, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('imagem após fechameto')
plt.imshow(closing, 'gray')
```

<p>Calculo das distâncias euclidianas das moedas através do método cv2.distanceTransform (1ª imagem abaixo) então é aplicado limiarização pra obter a posição dos marcadores (2ª imagem abaixo), com o valor do limiar controlamos um nível mais adequado :</p>

```python
dist = cv2.distanceTransform(closing, cv2.DIST_L2, 3)
ret, dist1 = cv2.threshold(dist, 0.3*dist.max(), 255, 0)
ret, dist2 = cv2.threshold(dist, 0.5*dist.max(), 255, 0)
ret, dist3 = cv2.threshold(dist, 0.7*dist.max(), 255, 0)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('distâncias')
plt.imshow(dist, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('nivel 01')
plt.imshow(dist1, 'gray')

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('nivel 02')
plt.imshow(dist2, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('nivel 03')
plt.imshow(dist3, 'gray')

```

<p></p>

<p>O trecho abaixo cria os marcadores necessarios para a aplicação do método cv2.watershed:</p>

```python
markers = np.zeros(dist.shape, dtype=np.int32)
dist_8u = dist3.astype('uint8')
contours, _ = cv2.findContours(dist_8u, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours)):
    cv2.drawContours(markers, contours, i, (i+1), -1)
    
markers = cv2.circle(markers, (15,15), 5, len(contours)+1, -1)

plt.figure(figsize=(12,8))
plt.title('Marcadores')
plt.imshow(markers, cmap='gray')
```

<p></p>

<p>Resultado da aplicação do metodo cv2.watershed com o nível 03 acima necessário para separar as moedas sobrepostas na imagem:</p>

```python
markers = cv2.watershed(img, markers)
img[markers == -1] = [255,0,0]

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.subplot(1,2,1).set_title('watershed')
plt.imshow(markers, 'gray')
plt.subplot(1,2,2)
plt.subplot(1,2,2).set_title('Resultado final')
plt.imshow(img, 'gray')
```
