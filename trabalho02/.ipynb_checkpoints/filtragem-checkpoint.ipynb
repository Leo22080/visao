<h1>1. Filtragem (espacial ou na frequência)</h1>

```python
import cv2
import numpy as np
import scipy.fftpack as fp
import matplotlib.pyplot as plt
```

<p>Bibliotecas utilizadas e função auxiliar para o calculo da função gaussiana da imagem com desvio sigma</p>

```python
def gauss(S, sigma=1, vmax= 0.05):
    '''
        Retorna uma gaussiana de média zero e desvio padrao sigma       
        G(u, v) = 1/(2*pi*sigma^2) * exp(-(u^2+v^2))/(2*sigma^2)
        no intervalo [-vmax, vmax]x[-vmax, vmax]
        deslocado com scipy.fftshift 
        retorna um array com mesma dimensão de S
    '''   
    
    Nu, Nv = S.shape
    u = Nu * np.linspace(-vmax, vmax, Nu)
    v = Nv * np.linspace(-vmax, vmax, Nv)
    U, V = np.meshgrid(v, u)
    
    G = np.exp(-(U*U + V*V) / 2.0 / sigma**2)
     
    G = fp.fftshift(G)
    
    return G / sigma**2
```

<h3>Carregando e mostrando a imágem original:</h3>

```python

img = cv2.imread('imgs/pinguins2.jpeg', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12,8))
plt.title('imagem original')
plt.imshow(img, cmap='gray')
#plt.colorbar()
```

<h3>Calculando a FFT 2D e tratando a FFT para os gráficos:</h3>
<p>Obtemos a transformada de Furier da imagem (FFT) atravez da função fft2 do módulo scipy.fftpack e gravado na variável F. A tranformada e então tratada e então mostrada em escala logarítmica.</p>

```python
F = fp.fft2(img) #Calculando a FFT 2D da imágem

#tratando a FFT para os gráficos
Fm = np.absolute(F)
Fm /= Fm.max()
Fm = fp.fftshift(Fm)
Fm = np.log(Fm) #FFT em escala logaritma

plt.figure(figsize=(12,8))
plt.title('FFT em escala log')
plt.imshow(Fm, cmap='gray')
```

<h3>Aplicando o desfoque (filtro) gaussiano:</h3>
<p>O filtro gaussiano do tamanho da imagem é gerado com desvio de 3, através da função gauss e gravado na variável G. O filtro é aplicado na FFT da imágem na variável Fg. E então devidamente tratado para ser exibido abaixo</p>

```python
sigma = 3 #desvio padrao
G = gauss(img, sigma)

Fg = F * G

plt.figure(figsize=(12,8))
plt.title('FFT filtrada em escala log')
Fga = np.absolute(Fg)
Fga = fp.fftshift(Fga+1e-6)
Fga = np.log(Fga)
plt.imshow(Fga, cmap='gray')
```

<h3>Mostrado o resultado:</h3>
<p>Obtemos a tranformada inversa atraves da função ifft2 de scipy.fftpack, a imagem e tratada e por fim exibida. Onde o desfoque da imagem apos a aplicação do filtro pode ser observado.</p>

```python
img_blur = fp.ifft2(Fg)
img_blur = np.absolute(img_blur)

plt.figure(figsize=(12,8))
plt.title('imagem filtrada')
plt.imshow(img_blur, cmap='gray')
```

<h3>Carregando e mostrando a outra imagem:</h3>

<p>Fazendo uma pequena modificação na função gauss, para que ele passe a filtrar as baixas frequências:</p>

```python
img2 = cv2.imread('imgs/turma-monica.jpg', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12,8))
plt.title('imagem original')
plt.imshow(img2, cmap='gray')
```

```python
def gauss_inv(S, sigma=1, vmax= 0.05):
    
    
    Nu, Nv = S.shape
    u = Nu * np.linspace(-vmax, vmax, Nu)
    v = Nv * np.linspace(-vmax, vmax, Nv)
    U, V = np.meshgrid(v, u)
    
    G = 1 - np.exp(-(U*U + V*V) / 2.0 / sigma**2)
     
    G = fp.fftshift(G)
    
    return G / sigma**2
```

```python
F = fp.fft2(img2) #Calculando a FFT 2D da segunda imágem

sigma = 3 #desvio padrao
G = gauss_inv(img2, sigma)

Fg = F * G

plt.figure(figsize=(12,8))
plt.title('FFT filtrada em escala log da segunda imagem')
Fga = np.absolute(Fg)
Fga = fp.fftshift(Fga+1e-6)
Fga = np.log(Fga)
plt.imshow(Fga, cmap='gray')
```

```python
saida = fp.ifft2(Fg)
saida = np.absolute(saida)

plt.figure(figsize=(12,8))
plt.title('imagem filtrada')
plt.imshow(saida, cmap='gray')
```
